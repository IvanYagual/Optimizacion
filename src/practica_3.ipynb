## 4.1 Optimizacion para una funci√≥n no deribable pero convexa

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def objective(v):
    x, y = v
    f = np.sqrt(x**2 + y**2)
    return f
```

```python
r_min, r_max = -1.0, 1.0
xaxis = np.arange(r_min, r_max, 0.1)
yaxis = np.arange(r_min, r_max, 0.1)
x, y = np.meshgrid(xaxis, yaxis)
v = x, y
result = objective(v)

fig, ax =  plt.subplots(subplot_kw={'projection':'3d'}, figsize=(30,30))
ax.plot_surface(x, y, result, cmap='jet')

plt.show()
```

```python
from scipy.optimize import minimize
```

```python
initial_guess =r_min + np.random.rand(2) * (r_max - r_min)

result = minimize(objective, initial_guess, method='powell')


print("Estado: %s" %result['message'])
print("N√±umero total de evaluaciones de la funcion objetivo: %d" %result['nfev'])
print("Solucion =", result['x'])

solution = result['x']
print("(solucion) =", objective(solution))
```

## 4.2 Optimizacion de una funcion derobable pero no convexa

```python
# funcion de Ackley
def ackley(v):
    x, y = v
    f = -20.0 * np.exp(-0,2 * np.sqrt(0.5 * (x**2 + y**2))) - np.exp(0.5 * (np.cos(2 * np.pi * x) + np.cos(2 * np.pi * y))) + np.e + 20
    return f
```

```python
r_min, r_max = -5.0, 5.0
xaxis = np.arange(r_min, r_max, 0.1)
yaxis = np.arange(r_min, r_max, 0.1)
x, y = np.meshgrid(xaxis, yaxis)
v = x, y
result = ackley(v)

fig, ax =  plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(x, y, result, cmap='jet')

plt.show()
```

```python
initial_guess =r_min + np.random.rand(2) * (r_max - r_min)

result = minimize(ackley, initial_guess, method='powell')


print("Estado: %s" %result['message'])
print("N√±umero total de evaluaciones de la funcion objetivo: %d" %result['nfev'])
print("Solucion =", result['x'])

solution = result['x']
print("(solucion) =", ackley(solution))
```
