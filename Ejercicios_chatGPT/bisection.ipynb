```python
def bisection_method_optimization(f, a, b, tol=1e-5, max_iter=1000):
    """
    Bisection method for finding the minimum of a unimodal function f in the interval [a, b].

    Parameters:
    f        : The function to optimize.
    a, b     : The interval [a, b] where the function is defined and unimodal.
    tol      : The tolerance level for convergence.
    max_iter : Maximum number of iterations.

    Returns:
    A tuple containing:
    - The approximate minimum point
    - The function value at the minimum
    - The number of iterations performed
    """

    iter_count = 0
    while (b - a) / 2 > tol and iter_count < max_iter:
        # Calculate the midpoint of the interval
        c = (a + b) / 2
        
        # Calculate the points near the midpoint for comparison
        d1 = a + (b - a) / 4
        d2 = b - (b - a) / 4
        
        # Evaluate the function at d1 and d2
        f_d1 = f(d1)
        f_d2 = f(d2)
        
        if f_d1 < f_d2:
            # The minimum is likely in the left half [a, d2]
            b = d2
        else:
            # The minimum is likely in the right half [d1, b]
            a = d1
        
        iter_count += 1

    # The midpoint will be our best estimate for the minimum
    min_point = (a + b) / 2
    min_value = f(min_point)

    return min_point, min_value, iter_count


# Example usage
if __name__ == "__main__":
    # Define a sample unimodal function (e.g., a parabola)
    def f(x):
        return (x - 2) ** 2 + 1  # Minimum at x = 2

    # Interval to search for the minimum
    a = 0
    b = 5

    # Call the bisection optimization method
    min_point, min_value, iterations = bisection_method_optimization(f, a, b)

    # Output the result
    print(f"Minimum point: {min_point}")
    print(f"Function value at minimum: {min_value}")
    print(f"Iterations: {iterations}")
```
